```python
#sierpinski gasket 
import matplotlib.pyplot as plt
import random as rd
```

```python
p1=[300,100]
p2=[200,300]
p3=[400,300]
plt.scatter(p1[0],p1[1],s=5)
plt.scatter(p2[0],p2[1],s=5)
plt.scatter(p3[0],p3[1],s=5)
seed_point=[100,150]
plt.scatter(seed_point[0],seed_point[1],s=5)
current_point=seed_point
for i in range(1000):
    choice=rd.choice([0,1,2])
    if choice==0:
        mid_point=[int((current_point[0]+p1[0])/2),int((current_point[1]+p1[1])/2)]
    elif choice==1:
        mid_point=[int((current_point[0]+p2[0])/2),int((current_point[1]+p2[1])/2)]
    elif choice==2:
        mid_point=[int((current_point[0]+p3[0])/2),int((current_point[1]+p3[1])/2)]
    current_point=mid_point
    #print(current_point)
    plt.scatter(current_point[0],current_point[1],s=5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random as rd
p1=[300,100]
p2=[200,300]
p3=[400,300]
seed_point=[200,150]
plt.scatter(seed_point[0],seed_point[1],s=5)
current_point=seed_point
for i in range(1000):
    p=rd.choice([0,1,2])
    if p==0:
        c=[(current_point[0]+p1[0])/2,(current_point[1]+p1[1])/2]
    if p==1:
        c=[(current_point[0]+p2[0])/2,(current_point[1]+p2[1])/2]
    if p==2:
        c=[(current_point[0]+p3[0])/2,(current_point[1]+p3[1])/2]
    current_point=c
    plt.scatter(current_point[0],current_point[1],s=5)
```
