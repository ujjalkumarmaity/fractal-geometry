```python
import matplotlib.pyplot as plt
```

```python
def midpoint(x,y):
    mid_x=int((x[1]-x[0])/3)
    mid_y=int((y[1]-y[0])/3)
    return [x[0]+mid_x,x[0]+(2*mid_x)],[y[0]+mid_y,2*mid_y+y[0]]

def plot_r(x,y):
    plt.plot([x[0],x[1]],[y[0],y[1]])
    plt.plot([x[0],x[2]],[y[0],y[2]])
    plt.plot([x[1],x[3]],[y[1],y[3]])
    plt.plot([x[2],x[3]],[y[2],y[3]])
def fun(x,y,count):
    if count>0:
        count-=1
        x1,y1=midpoint([x[0],x[1]],[y[0],y[1]])     #0 0  30 60
        x2,y2=midpoint([x[0],x[2]],[y[0],y[2]])   # 30 60  0 0 
        x3,y3=midpoint([x[1],x[3]],[y[1],y[3]])   # 30 60   90 90
        x4,y4=midpoint([x[2],x[3]],[y[2],y[3]])  #90 90   30 60

        plot_r([x1[0],x1[0],x2[0],x2[0]],[y2[0],y1[0],y2[0],y1[0]])
        fun([x1[0],x1[0],x2[0],x2[0]],[y2[0],y1[0],y2[0],y1[0]],count)
        plot_r([x2[0],x2[0],x2[1],x2[1]],[y2[0],y1[0],y2[0],y1[0]])
        fun([x2[0],x2[0],x2[1],x2[1]],[y2[0],y1[0],y2[0],y1[0]],count)
        plot_r([x2[1],x2[1],x[3],x[3]],[y2[0],y1[0],y2[0],y1[0]])
        fun([x2[1],x2[1],x[3],x[3]],[y2[0],y1[0],y2[0],y1[0]],count)

        
        plot_r([x1[0],x1[0],x2[0],x2[0]],[y1[0],y1[1],y1[0],y1[1]])
        fun([x1[0],x1[0],x2[0],x2[0]],[y1[0],y1[1],y1[0],y1[1]],count)
        plot_r([x1[0],x1[0],x2[0],x2[0]],[y1[1],y3[1],y1[1],y3[1]])
        fun([x1[0],x1[0],x2[0],x2[0]],[y1[1],y3[1],y1[1],y3[1]],count)

        plot_r([x2[1],x2[1],x[3],x[3]],[y1[0],y1[1],y1[0],y1[1]])
        fun([x2[1],x2[1],x[3],x[3]],[y1[0],y1[1],y1[0],y1[1]],count)
        plot_r([x2[1],x2[1],x[3],x[3]],[y1[1],y3[1],y1[1],y3[1]])
        fun([x2[1],x2[1],x[3],x[3]],[y1[1],y3[1],y1[1],y3[1]],count)

        plot_r([x2[0],x2[0],x2[1],x2[1]],[y1[1],y3[1],y1[1],y3[1]])
        fun([x2[0],x2[0],x2[1],x2[1]],[y1[1],y3[1],y1[1],y3[1]],count)
    
x=[0,0,90,90]
y=[0,90,0,90]
plot_r(x,y)
fun(x,y,3)
plt.show()

```
