```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

# base: devided line into two eqal part and then rotate upper part of line

```python
n=0
def midpoint(x,y):
    return [(x[0]+x[1])/2,(y[0]+y[1])/2]

def rotate_around_point(n,mid_point, radians,x1,y1):
    x=[mid_point[0],x1[1]]
    y=[mid_point[1],y1[1]]
    x_point=(x[0]+(math.cos(radians))*(x[1]-x[0]) - math.sin(radians)*(y[1]-y[0]))
    y_point=(y[0]+math.sin(radians)*(x[1]-x[0]) + math.cos(radians)*(y[1]-y[0]))
    x_point1=(x[0]+(math.cos(-radians))*(x[1]-x[0]) - math.sin(-radians)*(y[1]-y[0]))
    y_point1=(y[0]+math.sin(-radians)*(x[1]-x[0]) + math.cos(-radians)*(y[1]-y[0]))

    plt.plot(x1,y1) #main  line
    plt.plot([x[1],x_point],[y[1],y_point])
    plt.plot([x[1],x_point1],[y[1],y_point1])

    mid1=midpoint([x[1],x_point],[y[1],y_point])
    

    mid2=midpoint([x[1],x_point1],[y[1],y_point1])

    if n<7:
        rotate_around_point(n+1,mid1,radians,[x[1],x_point],[y[1],y_point])
        rotate_around_point(n+1,mid2,-radians,[x[1],x_point1],[y[1],y_point1])

x=[0,0]
y=[0,10]
q=midpoint(x,y)
rotate_around_point(n+1,q,math.radians(65),x,y)
```
