```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def logistic(x,r):
    xn=[]
    itration=[]
    for i in range(0,100):
        x=r*x*(1-x)
        xn.append(x)
        itration.append(i)
    plt.plot(itration,xn)
    plt.show()
```

# cobweb

```python
import numpy as np
t = np.linspace(0, 1,100)
def logistic(x,r):
    xn=[]
    itration=[]
    for i in range(0,100):
        xi=r*x*(1-x)
        xn.append(x)
        itration.append(xi)
        x=xi
    return xn
```

```python
a=[]
for i in range(100):
    a.append(4*t[i]*(1-t[i]))
plt.plot(t,a)
plt.plot(t,t)
plt.plot([0,0],[0,1])
plt.plot([1,0],[0,0])
plt.plot([1,1],[1,0])
plt.plot([0,1],[1,1])
xn=logistic(.5,3)
plt.plot([xn[0],xn[0]],[0,xn[1]])
for i in range(98):
    plt.plot([xn[i],xn[i+1]],[xn[i+1],xn[i+1]])
    plt.plot([xn[i+1],xn[i+1]],[xn[i+1],xn[i+2]])

plt.show()
```
