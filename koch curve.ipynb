```python
import matplotlib.pyplot as plt
import math
def rotate(x,y):
    x_point=(x[0]+(math.cos(math.radians(60)))*(x[1]-x[0]) - math.sin(math.radians(60))*(y[1]-y[0]))
    y_point=(y[0]+math.sin(math.radians(60))*(x[1]-x[0]) + math.cos(math.radians(60))*(y[1]-y[0]))
    return x_point,y_point

```

```python

def cantor(x,y,count):
    if count>0:
        count-=1
        mid_x1=(2*x[0]+x[1])/3
        mid_x2=(x[0]+x[1]*2)/3
        mid_y1=(2*y[0]+y[1])/3
        mid_y2=(y[0]+y[1]*2)/3
        
        
        new_x,new_y=rotate([mid_x1,mid_x2],[mid_y1,mid_y2])
        
        #print([x[0],mid_x1],[y[0],mid_y1])               
        cantor([x[0],mid_x1],[y[0],mid_y1],count)
        cantor([mid_x2,x[1]],[mid_y2,y[1]],count)
        cantor([mid_x1,new_x],[mid_y1,new_y],count)
        cantor([new_x,mid_x2],[new_y,mid_y2],count)
        if count==1:
            plt.plot([x[0],mid_x1],[y[0],mid_y1])
            plt.plot([mid_x2,x[1]],[mid_y2,y[1]])

            plt.plot([mid_x1,new_x],[mid_y1,new_y])
            plt.plot([new_x,mid_x2],[new_y,mid_y2])

def tri(x,y,c):
    cantor([x[0],x[1]],[y[0],y[1]],c)
    cantor([x[1],x[2]],[y[1],y[2]],c)
    cantor([x[2],x[0]],[y[2],y[0]],c)

x=[0,6,3]
y=[3,3,0]
tri(x,y,5)
plt.show()
```
