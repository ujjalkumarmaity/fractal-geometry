```python
import matplotlib.pyplot as plt
import random as rd
```

```python
x=[100,200,300]
y=[100,200,100]
plt.plot([x[0],x[1]],[y[0],y[1]])
plt.plot([x[1],x[2]],[y[1],y[2]])
plt.plot([x[0],x[2]],[y[0],y[2]])
```

```python
def tri(x,y,count):
    if count!=0:
        mid=[int((x[0]+x[1])/2),int((x[1]+x[2])/2),int((x[0]+x[2])/2)]
        mid1=[int((y[0]+y[1])/2),int((y[1]+y[2])/2),int((y[0]+y[2])/2)]
        #print(x,y)
        x1=[x[0],mid[0],mid[2]]
        y1=[y[0],mid1[0],mid1[2]]
        x2=[mid[0],x[1],mid[1]]
        y2=[mid1[0],y[1],mid1[1]]
        x3=[mid[2],mid[1],x[2]]
        y3=[mid1[2],mid1[1],y[2]]
        plt.plot([x1[0],x1[1]],[y1[0],y1[1]])
        plt.plot([x1[1],x1[2]],[y1[1],y1[2]])
        plt.plot([x1[0],x1[2]],[y1[0],y1[2]])
        
        plt.plot([x2[0],x2[1]],[y2[0],y2[1]])
        plt.plot([x2[1],x2[2]],[y2[1],y2[2]])
        plt.plot([x2[0],x2[2]],[y2[0],y2[2]])
        
        plt.plot([x3[0],x3[1]],[y3[0],y3[1]])
        plt.plot([x3[1],x3[2]],[y3[1],y3[2]])
        plt.plot([x3[0],x3[2]],[y3[0],y3[2]])
 
 
        count-=1
        tri(x1,y1,count)
        tri(x2,y2,count)
        tri(x3,y3,count)
        
    else:
        return 0;
    
plt.plot([x[0],x[1]],[y[0],y[1]])
plt.plot([x[1],x[2]],[y[1],y[2]])
plt.plot([x[0],x[2]],[y[0],y[2]])

tri(x,y,5)
plt.show()
```
