```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def tent(x,r):
    xn=[]
    itration=[]
    for i in range(0,100):
        if 0<=x<=.5:
            x=2*r*x
        elif .5<x<=1:
            x=r*2*(1-x)
        xn.append(x)
        itration.append(i)
    plt.plot(itration,xn)
    plt.show()
```

```python
tent(.5,.2)
```

```python
tent(.5,.51)
```

```python
tent(.5,.65)
```

```python
tent(.5,.75)
```
